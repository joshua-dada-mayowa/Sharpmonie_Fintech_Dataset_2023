```python
import pandas as pd
import numpy as np
from multiprocessing import Pool

def process_sender_row(sender_row):
    sender_account_id = sender_row['AccountID']
    sender_transaction_date = sender_row['TransactionDate']
    sender_amount = sender_row['Amount']
    transaction_id = sender_row['TransactionID']
    status = sender_row['Status']
    payment = sender_row['PaymentMethod']

    # Check if there are potential receivers for the sender
    if sender_account_id in receiver_data:
        receiver_row = receiver_data[sender_account_id]
        receiver_wallet_id = receiver_row['AccountID']
        receiver_transaction_date = receiver_row['TransactionDate']
        receiver_amount = receiver_row['Amount']

        return [
            p2p_id, transaction_id, sender_account_id, receiver_wallet_id,
            sender_amount, status, payment, sender_transaction_date
        ], [
            p2p_id, transaction_id, receiver_wallet_id, sender_account_id,
            receiver_amount, status, payment, receiver_transaction_date
        ]

    return None, None

# Rest of your code remains the same to prepare p2p_df, potential_receivers, and receiver_data

p2p_data = []
p2p_id = 1

# Define the number of processes to run in parallel
num_processes = 4  # You can adjust this based on your system's capabilities

# Split p2p_df into chunks to be processed by different processes
chunk_size = len(p2p_df) // num_processes
chunks = [p2p_df[i:i + chunk_size] for i in range(0, len(p2p_df), chunk_size)]

# Process chunks in parallel
with Pool(num_processes) as pool:
    results = pool.map(process_sender_row, chunks)

# Collect results from the parallel processing
for result_set in results:
    for sender_data, receiver_data in result_set:
        if sender_data and receiver_data:
            p2p_data.extend([sender_data, receiver_data])
            p2p_id += 1

# Create the P2P DataFrame directly from the list
p2p_columns = ['P2PID', 'TransactionID', 'SenderAccountID', 'ReceiverAccountID',
               'Amount', 'Status', 'PaymentMethod', 'TransactionDate']

p2p_df = pd.DataFrame(p2p_data, columns=p2p_columns)
```
